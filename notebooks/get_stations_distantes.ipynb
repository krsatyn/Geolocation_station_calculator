```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
station_3 = []
station_4 = []
station_5 = []
station_6 = []


station_3_time = []
station_4_time = []
station_5_time = []
station_6_time = []



with open(r"src\station_dist\distances.txt", "r", encoding="utf-8") as log:
    for line in log:        
        # print(line.split(' ')[1].split(';'))
        if line.split(' ')[1].split(';')[1] == '3':
            station_3.append(float(line.split(' ')[1].split(';')[2]))
            station_3_time.append(line.split(' ')[1].split(';')[0])
            
        elif line.split(' ')[1].split(';')[1] == '4':
            station_4.append(float(line.split(' ')[1].split(';')[2]))
            station_4_time.append(line.split(' ')[1].split(';')[0])
            
        elif line.split(' ')[1].split(';')[1] == '5':
            station_5.append(float(line.split(' ')[1].split(';')[2]))
            station_5_time.append(line.split(' ')[1].split(';')[0])
            
        else:
            station_6.append(float(line.split(' ')[1].split(';')[2]))
            station_6_time.append(line.split(' ')[1].split(';')[0])
```

```python
print(f'{len(station_3)=}')
print(f'{len(station_4)=}')
print(f'{len(station_5)=}')
print(f'{len(station_6)=}')
```
